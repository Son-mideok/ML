### 2. 데이터 준비

```python
# 원핫 인코딩 직접만들기 
import numpy as np

def vectorize_sequences(seauences, dimension=10000):
    # print(len(sequences))
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = -1. # i행 첫번째 리뷰, 
    return results

# 훈련 데이터를 벡터로 변환
x_train = vectorize_sequences(train_data).astype('float32')

# seauences : 벡터화된 데이터 train data
```

```python
y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')
```

```python
# 블로그

import numpy as np
def vectorize_sequence(sequences, dimension=10000):
  results = np.zeros((len(sequences), dimension))
  for idx, sequence in enumerate(sequences):
      results[idx, sequence] = 1.
  return results

x_train = vectorize_sequence(train_data)
x_test = vectorize_sequence(test_data)

print(x_train.shape, x_test.shape)
```

### 3. 신경망 모델 만들기

```python
![3-layer network](https://s3.amazonaws.com/book.keras.io/img/ch3/3_layer_network.png)
```

```python
from tensorflow.keras import models 
from tensorflow.keras import layers

model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape = (10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
# 모델 compile 셋팅

from tensorflow.keras import models
from tensorflow.keras import layers
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
2:54
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

### 4. 훈련검증

```python
x_val = x_train[:10000].astype(float)
partial_x_train = x_train[10000:].astype(float)
y_val = y_train[:10000].astype(float)
partial_y_train = y_train[10000:].astype(float)
```

```python
history = model.fit(partial_x_train,
                   partial_y_train,
                   epochs = 20,
                   batch_size = 512,
                   validation_data=(x_val, y_val))
```
